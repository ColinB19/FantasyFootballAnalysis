This is the section for the polished process

```python
class DataPipeline():

    def __init__(self):

        pass

    def load_data(self, path = "./data"):
        self.wr_stats = pd.read_csv(f'{path}/agg_wr_final.csv')
        self.def_points_allowed = pd.read_csv(f'{path}/def_fpoints.csv').rename({'defteam':'team'}, axis=1)
        self.def_injuries = pd.read_csv(f'{path}/inj_defense.csv')
        self.qb_stats = pd.read_csv(f'{path}/qb_stats.csv')
    
    def preprocess(self):

        self.__pre_check()
        self.__clean_data()
        self.__create_windowed_data()
        self.__post_check()


    def __pre_check(self):
        pass

    def __post_check(self):
        pass

    def __clean_data(self):
        
        dataset_length_pre_clean = len(self.wr_stats)
        # drop columns with a lot of missing data
        self.wr_stats.drop(['avg_cushion', 
               'avg_separation', 
               'twitter_username', 
               'receiving_drop', 
               'receiving_broken_tackles',
               'receiving_rat',
               'receiving_int',
               'receiving_drop_pct',
               'draft_round',
               'draft_pick',
               'draft_ovr',
               'target_share_4',
               'snap_percentage_4'], axis=1, inplace = True)

        # there are a few of these and these players are not very involved in offense
        self.wr_stats.drop(self.wr_stats[self.wr_stats.wopr.isna()].index, inplace=True)

        # clean up any infinities
        self.wr_stats.replace([np.inf, -np.inf], 0, inplace=True)

        # fill in the depth with the mode for that player
        self.wr_stats["depth_team"] = self.wr_stats.groupby("player_id")["depth_team"].transform(
            lambda x: x.fillna(x.mode()[0]) if len(x.mode()) > 0 else x.fillna(np.nan)
        )
        # the rest of these players are low on the depth chart so let's just fill with the max
        self.wr_stats["depth_team"] = self.wr_stats["depth_team"].fillna(self.wr_stats.depth_team.max())

        # for the rest of the null stats, only 18 players have this issue. Of these, none are very impactful players (from a fantasy football persepctive). We will be dropping these statlines
        self.wr_stats.drop(self.wr_stats[self.wr_stats.college.isna()].index, inplace = True)

        # we're also going to drop all 0,0,0 statlines
        self.wr_stats.drop(self.wr_stats[(self.wr_stats['receptions'] == 0) & (self.wr_stats['total_targets'] == 0) & (self.wr_stats['receiving_yards'] == 0)].index, inplace = True)
        
        dataset_length_post_clean = len(self.wr_stats)

        print(f"The cleaning process removed {dataset_length_pre_clean - dataset_length_post_clean} rows from our WR dataset.")

    def visualize_data(self, **kwargs):
        try:
            r = kwargs['plt_rows']
        except:
            r = 1
        
        try:
            c = kwargs['plt_cols']
        except:
            c = 1
        fig, ax = plt.subplots(r, c, figsize = (10,10))

        sns.histplot(self.wr_stats[self.wr_stats['receiving_fpoints'] > 0]['receiving_fpoints'], kde=True, ax=ax)
        plt.show()

    def __create_windowed_data(self, roll_window = 2, pos_subset = 'WR'):
        # now we need to get the dataset to be averages of all datapoints aside from fantasy points before the current week.
        # subset to just a single position
        sub_stats = self.wr_stats[self.wr_stats['position'] == pos_subset]
        dataset_length_pre_window = len(sub_stats)
        # ensure the data is sorted
        sub_stats.sort_values(by=['player_id', 'week'], inplace=True)

        # meta info for each player/game that we want to predict on! (So we can merge past game info onto this)
        game_keys = sub_stats[['player_id', 'game_id', 'team', 'week', 'position', 'receiving_fpoints', 'age', 'height', 'weight', 'depth_team', 'opp_team']]

        # this is what will become our historic data to merge onto game_keys
        prev_stats = sub_stats[['player_id', 'game_id', 'receiving_yards', 'avg_yac', 'receptions',
            'receiving_touchdowns', 'week', 'targets_1', 'targets_2', 'targets_3',
            'targets_4', 'total_targets', 'rz_targets', 'garbage_time_fpoints',
            'receiving_fpoints', 'avg_depth_of_target', 'air_yards',
            'max_target_depth', 'fumble_lost', 'receiving_first_downs',
            'receiving_epa', 'receiving_2pt_conversions', 'unrealized_air_yards',
            'racr', 'snap_count_1', 'snap_count_2', 'snap_count_3', 'snap_count_4',
            'total_relevant_snaps', 'snap_percentage_1',
            'snap_percentage_2', 'snap_percentage_3', 'snap_percentage',
            'target_share_1', 'target_share_2', 'target_share_3', 'target_share',
            'air_yards_share', 'wopr']]
        
        # now we get the rolling average for each player each week.
        prev_stats = prev_stats.set_index(['game_id', 'week']).groupby(['player_id']).rolling(roll_window, min_periods=1).mean().reset_index().rename({'receiving_fpoints':'past_fpoints'}, axis=1)

        # note that not all players play each week. So we need to be careful about which rolling window average stats we associate with which prediction week.
        # e.g. if we want to predict a players fantasy output after coming back from injury, we want that players stats from before they got injured, not the last two weeks where they were injured.
        for _, row in game_keys.iterrows():
            # grab the week and player_id for each prediction row
            week = row.week
            pid = row.player_id
            # associate that row with the most recent available stats for that player
            temp_week = prev_stats[(prev_stats.player_id == pid) & (prev_stats.week < week)].week.max()
            index = prev_stats[(prev_stats.player_id == pid) & (prev_stats.week == temp_week)].index
            prev_stats.loc[index, 'new_week'] = week


        # now we merge cma stats
        self.data = game_keys.merge(prev_stats.drop('game_id', axis=1), left_on=['player_id', 'week'], right_on=['player_id', 'new_week'], suffixes = ('', '_remove'))
        self.data.drop(['new_week'] + [x for x in self.data.columns if '_remove' in x], axis=1, inplace = True)
        dataset_length_post_window = len(self.data)
        print(f"The windowing process removed {dataset_length_pre_window - dataset_length_post_window} rows from our WR dataset.")

        # finally, we are going to take the top-40 performances every week, to prevent the model from being 0-biased
        # FIXME: The windowing is messing this up. A player may not have two top-40 performances in a row.
        self.data.sort_values(by=['week', 'receiving_fpoints'], ascending = [True, False], inplace = True)
        self.data = self.data.groupby('week').head(40)
        
    def get_test_train_data(self):
        self.data["week"] = self.data["week"] / 18
        self.data["depth_team"] = self.data["depth_team"] / 3
        drop_list = ['receiving_fpoints', 'player_id', 'game_id', 'team', 'opp_team', 'position']
        X_train, X_test, y_train,y_test = train_test_split(self.data.drop(drop_list, axis=1), 
                                                                                self.data['receiving_fpoints'], 
                                                                                test_size = 0.33, 
                                                                                random_state = 42)
        
        # scaling
        non_scale_list = [
                            "week",
                            "depth_team",
                            "snap_percentage_1",
                            "snap_percentage_2",
                            "snap_percentage_3",
                            "target_share",
                            "air_yards_share",
                            "target_share_1",
                            "target_share_2",
                            "target_share_3",
                            "target_share_1",
                            "wopr"
                        ]
        # columns you don't need to scale:
        scale_list = [
            x
            for x in self.data.columns
            if x
            not in non_scale_list + drop_list
        ]


        self._feature_list = scale_list + non_scale_list

        # scale train data
        self.X_train = X_train[scale_list]
        leftover_train = np.asarray(
                    X_train[non_scale_list]
                )
        self.y_train = np.asarray(y_train)

        train_scaler = StandardScaler()
        self.X_train_s = train_scaler.fit_transform(X=self.X_train)
        self.X_train_s_all = np.concatenate((self.X_train_s, leftover_train), axis=1)

        # scale test data
        self.X_test = X_test[scale_list]
        leftover_test = np.asarray(
                    X_test[non_scale_list]
                )
        self.y_test = np.asarray(y_test)

        test_scaler = StandardScaler()
        self.X_test_s = test_scaler.fit_transform(X=self.X_test)
        self.X_test_s_all = np.concatenate((self.X_test_s, leftover_test), axis=1)

        print(f"shape1 {self.X_train_s.shape}, shape2: {self.X_train_s_all.shape}")
```

```python
dpl = DataPipeline()
dpl.load_data()
dpl.preprocess()
dpl.get_test_train_data()
```

```python
# none of these correlations are particularly strong. But we may want to filter out some of this data anyway. 
# let's train a singular regression model on each far and see what happens
d = []
for idx, var in enumerate(dpl._feature_list):
    temp = {}
    X_train_sub = dpl.X_train_s_all[:, idx]
    scores = cross_val_score(reg, X_train_sub.reshape(-1, 1), dpl.y_train, cv=3, scoring='neg_root_mean_squared_error')
    temp['Feature'] = var
    temp['score_0'] = scores[0]
    temp['score_1'] = scores[1]
    temp['score_2'] = scores[2]
    temp['average_score'] = scores.mean()

    d.append(temp)

single_regression_outcomes = pd.DataFrame(d).sort_values('average_score', ascending=False)
```

```python
# so we expect wopr, past_fpoints, first_downs, etc to be in the model, let's automate this process
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

# feature selection
def select_features(X_train, y_train, X_test):
    # configure to select all features
    fs = SelectKBest(score_func=f_regression, k=7)
    # learn relationship from training data
    fs.fit(X_train, y_train)
    # transform train input data
    X_train_fs = fs.transform(X_train)
    # transform test input data
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs

X_train_trans, X_test_trans, fs = select_features(dpl.X_train_s_all, dpl.y_train, dpl.X_test_s_all)

# for i in range(len(fs.scores_)):
#     print(f'Score, Feature {dpl._feature_list[i]}: {fs.scores_[i]}\nP-Value, Feature {dpl._feature_list[i]}: {fs.pvalues_[i]}')

X_train_trans.shape
```
